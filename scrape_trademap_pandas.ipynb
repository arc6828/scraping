```python
import pandas as pd
import os
```

```python
html_files = os.listdir("html")
html_files
```

```python
paths = [os.path.join("html", file) for file in html_files]
paths
files = [ { "filename" : path,  "size" : int (os.stat(path).st_size / 1024) } for path in paths]
files
```

```python
#http://localhost:8888/view/sample.html
#dfs = pd.read_html('html/E_BE_100610.html',header=[0])
filename = html_files[1]
dfs = pd.read_html('html/'+filename, header=[0])
dfs[0]
```

```python
import json
df = dfs[0]
result = df.to_json(orient="split")
print(result)
result = json.loads(result)
```

```python
data_list = []
heads = result["columns"]
data = result["data"]

for cols in data : 
    #print(cols)
    if len(cols) < 2 :
        continue
        
    for head ,  col in zip(heads ,  cols):
        if not str(col).isnumeric() :
            continue
        if cols[1] == "World" :
            continue
        print(head.replace("M","")+"-01", cols[1], col, "kg" )
        data_list.append({
            "date" : head.replace("M","")+"-01" ,
            "country" : cols[1],
            "value" : col,
            "unit" : "kg",
        })
        







    
```

```python
import json

with open('data/'+filename+'.json', 'w') as outfile:
    json.dump(data_list, outfile)
```
